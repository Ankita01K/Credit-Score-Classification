### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Type_of_Loan']=le.fit_transform(df['Type_of_Loan'])
df
```

```python
numcol = df[['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
       'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Type_of_Loan',
       'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit',
       'Num_Credit_Inquiries', 'Credit_Mix', 'Outstanding_Debt',
       'Credit_Utilization_Ratio', 'Credit_History_Age',
       'Payment_of_Min_Amount', 'Total_EMI_per_month',
       'Amount_invested_monthly', 'Monthly_Balance', 'Credit_Score',
       'Payment_Behaviour_High_spent_Medium_value_payments',
       'Payment_Behaviour_High_spent_Small_value_payments',
       'Payment_Behaviour_Low_spent_Large_value_payments',
       'Payment_Behaviour_Low_spent_Medium_value_payments',
       'Payment_Behaviour_Low_spent_Small_value_payments']]

for i in numcol.columns:
    sns.displot(numcol[i])

    plt.title(f'Distribution of {i}')
    plt.show()
```

```python
for c in numcol:
    #getting upper lower quartile values
    q25,q75=np.percentile(df[c],25),np.percentile(df[c],75)
    IQR=q75-q25
    Threshold=IQR*1.5
    lower,upper=q25-Threshold,q75+Threshold
    Outliers=[i for i in df[c] if i < lower or i > upper]
    print('{} Total Number of outliers in {} Before Imputing : {}'.format('\033[1m',c,len(Outliers)))
    print('\n')
    #taking mean of a column without considering outliers
    df_include = df.loc[(df[c] >= lower) & (df[c] <= upper)]
    mean=int(df_include[c].mean())
    print('{} Mean of {} is {}'.format('\033[1m',c,mean))
    print('\n')
    #imputing outliers with mean
    df[c]=np.where(df[c]>upper,mean,df[c])
    df[c]=np.where(df[c]<lower,mean,df[c])
    Outliers=[i for i in df[c] if i < lower or i > upper]
    print('{} Total Number of outliers in {} After Imputing : {}'.format('\033[1m',c,len(Outliers)))
    print('\n')
```

# Defining dependent and independent variables

```python
X=df.loc[:,['Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
       'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Type_of_Loan',
       'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit',
       'Num_Credit_Inquiries', 'Credit_Mix', 'Outstanding_Debt',
       'Credit_Utilization_Ratio', 'Credit_History_Age',
       'Payment_of_Min_Amount', 'Total_EMI_per_month',
       'Amount_invested_monthly', 'Monthly_Balance', 'Credit_Score',
       'Payment_Behaviour_High_spent_Medium_value_payments',
       'Payment_Behaviour_High_spent_Small_value_payments',
       'Payment_Behaviour_Low_spent_Large_value_payments',
       'Payment_Behaviour_Low_spent_Medium_value_payments',
       'Payment_Behaviour_Low_spent_Small_value_payments']]
Y=df['Credit_Score']
```

# Splitting Training and Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, stratify = Y, random_state=3)
print(X.shape, X_train.shape, X_test.shape)
```

## **Feature Scalling**

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python

X_train = pd.DataFrame(X_train, columns = X.columns)

X_test = pd.DataFrame(X_test, columns=X.columns)
```

```python
X_train
```

##  **Class Distributions**

```python
from imblearn.over_sampling import SMOTE

 #Instantiate SMOTE
sm = SMOTE(random_state=42)

# Fit SMOTE to training data
X_train_res, Y_train_res = sm.fit_resample(X_train, Y_train)

# Print class distribution of original and resampled data
print('Class distribution before resampling:', Y_train.value_counts())
print('Class distribution after resampling:', Y_train_res.value_counts())
```


## Model Building

List of the models which I'm going used :

1-Logistic Regression.

2-Support Vector Classifier.

3-k-Nearest Neighbors Classifier.

4-Random Forest Classifier.

5-XGBoost.

## **Logistic Regression**

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import metrics

clf= LogisticRegression(max_iter=1000)
clf.fit(X_train_res, Y_train_res)

# Prediction
y_pred = clf.predict(X_test)
acc1 = accuracy_score(Y_test, y_pred)
print("Logistic Regression model accuracy (in %):", acc1*100)
print(classification_report(Y_test, y_pred))
```

```python
cf_matrix = metrics.confusion_matrix(Y_test,y_pred)
cf_matrix
sns.heatmap(cf_matrix, annot=True)
```



## **Support Vector Classifier**

```python
svc = SVC(kernel='linear')

# fit classifier to training set
svc.fit(X_train_res,Y_train_res)

# make predictions on test set
Y_pred=svc.predict(X_test)

# compute and print accuracy score
acc2 = accuracy_score(Y_test, Y_pred)
print("Support Vector Classifier model accuracy (in %):", acc2*100)
print(classification_report(Y_test, Y_pred))
```

```python
cf_matrix = metrics.confusion_matrix(Y_test,Y_pred)
cf_matrix
sns.heatmap(cf_matrix, annot=True)
```

# **k-Nearest Neighbors Classifier**

```python
classifier = KNeighborsClassifier(n_neighbors =5, metric = 'minkowski', p = 2)
classifier.fit(X_train_res, Y_train_res)

pred= classifier.predict(X_test)
acc3 = accuracy_score(Y_test, pred)
print("KNeighborsClassifier model accuracy (in %):", acc3*100)
print(classification_report(Y_test,pred))
```

```python
cf_matrix = metrics.confusion_matrix(Y_test,pred)
cf_matrix
sns.heatmap(cf_matrix, annot=True)
```

## **Random Forest Classifier**

Random Forest Classifier model with parameter n_estimators=1000

```python
rfc = RandomForestClassifier(n_estimators=1000,random_state=42)

# fit the model

rfc.fit(X_train_res, Y_train_res)

# Predict the Test set results

predictor = rfc.predict(X_test)
acc4 = accuracy_score(Y_test, predictor)
print("Model accuracy score with 100 decision-trees :", acc4*100)
print(classification_report(Y_test, predictor))
```

```python
cf_matrix = metrics.confusion_matrix(Y_test,predictor)
cf_matrix
sns.heatmap(cf_matrix, annot=True)
```

## **XGBoost**

```python
from xgboost import XGBClassifier
xg= XGBClassifier()
xg.fit(X_train_res, Y_train_res)
Prediction = xg.predict(X_test)
acc5 = accuracy_score(Y_test, Prediction)

print("Model accuracy score with 100 decision-trees :", acc5*100)
print(classification_report(Y_test, Prediction))
```

```python
cf_matrix = metrics.confusion_matrix(Y_test,Prediction)
cf_matrix
sns.heatmap(cf_matrix, annot=True)
```
